### Identify Common Issues in Tickets ###

```python
Most tickets mention products (16,607) and purchases (14,351), indicating frequent issues with product functionality, order status, refunds, or warranties. Common terms like issue (11,170), problem (2,168), error (1,300), and software (1,551) suggest customers face technical difficulties, compatibility problems, or system errors. Many customers seek assistance (6,255), troubleshooting steps (1,390), or data-related help (1,603), highlighting the need for better self-service options and automated support.
```

```python
import nltk
import re
from collections import Counter
import matplotlib.pyplot as plt
import seaborn as sns

# Function to clean text (same as you shared)
def clean_text(text):
    text = str(text).lower()
    text = re.sub('[({})?/$#|=]', '', text)
    text = re.sub('\n', '', text)
    text = re.sub('[.]', ' ', text)
    text = re.sub('[->]', ' ', text)
    text = re.sub('[:]', ' ', text)
    text = re.sub('[_]', ' ', text)
    text = re.sub('[,]', ' ', text)
    text = re.sub('[-]', ' ', text)
    text = re.sub('[\']', ' ', text)
    text = re.sub('<.*?>+', '', text)
    text = re.sub('\w*\d\w*', '', text)
    return text

# Apply text cleaning to Ticket Description
df_cleaned['Ticket Description'] = df_cleaned['Ticket Description'].apply(lambda x: clean_text(x))

# Remove stopwords (similar to your code)
from nltk.corpus import stopwords
nltk.download('stopwords')

stop_words = set(stopwords.words('english'))
stop_words.update(['i\'m', 'i\'ve', 'would', 'could', '-d', '-i\'m', '[--]:', 'not', 'mm', 'ca', 'cs', 'hi'])

def remove_stopwords(text):
    return [word for word in text.split() if word not in stop_words]

# Tokenize and clean Ticket Description
df_cleaned['temp1'] = df_cleaned['Ticket Description'].apply(lambda x: remove_stopwords(x))

# Count the most common words
top_words = Counter([word for sublist in df_cleaned['temp1'] for word in sublist])

# Create a DataFrame of the most common words
top_words_df = pd.DataFrame(top_words.most_common(20), columns=['Common Words', 'Count'])

# Plot the top 20 common words
plt.figure(figsize=(15,6))
sns.barplot(x='Count', y='Common Words', data=top_words_df)
plt.title('Top 20 Most Common Words in Ticket Descriptions')
plt.show()
```

The bar chart highlights that "product," "purchased," "issue," and "assist" are the most frequent words in ticket descriptions, indicating many support requests relate to product purchases and assistance. Words like "problem," "error," and "unable" suggest frequent technical difficulties, while "account," "device," and "update" point to login and software-related issues. To improve support, companies could implement self-service options, enhance product documentation, and prioritize assistance requests for faster resolution.

### Common Words in Ticket Subjects ###

```python
df_cleaned['Ticket Subject Tokens'] = df_cleaned['Ticket Subject'].apply(lambda x: remove_stopwords(x))

top_words_subj = Counter([word for sublist in df_cleaned['Ticket Subject Tokens'] for word in sublist])

top_words_subj_df = pd.DataFrame(top_words_subj.most_common(20), columns=['Common Words', 'Count'])

plt.figure(figsize=(15,6))
sns.barplot(x='Count', y='Common Words', data=top_words_subj_df)
plt.title('Top 20 Most Common Words in Ticket Subjects')
plt.show()
```

The bar chart reveals that "product," "issue," and "problem" are the most frequent words in ticket subjects, highlighting product-related concerns. Words like "compatibility," "refund," "bug," and "hardware" suggest common issues with technical problems and refund requests. Additionally, frequent terms such as "network," "installation," "setup," and "support" indicate recurring technical and setup challenges, while "account" and "payment" point to login and transaction-related concerns.

```python
import pandas as pd

# Convert time columns to datetime
df_cleaned["First Response Time"] = pd.to_datetime(df_cleaned["First Response Time"], errors='coerce')
df_cleaned["Time to Resolution"] = pd.to_datetime(df_cleaned["Time to Resolution"], errors='coerce')

# Ensure valid timestamps where resolution time is after response time
valid_times = df_cleaned["Time to Resolution"] > df_cleaned["First Response Time"]
df_cleaned = df_cleaned[valid_times]

# Calculate response time in hours
df_cleaned["Response Time (Hours)"] = (df_cleaned["Time to Resolution"] - df_cleaned["First Response Time"]).dt.total_seconds() / 3600

# Average response times
avg_first_response = (df_cleaned["First Response Time"] - df_cleaned["First Response Time"].min()).dt.total_seconds().mean() / 3600
avg_resolution_time = df_cleaned["Response Time (Hours)"].mean()

print(f" Average First Response Time: {avg_first_response:.2f} hours")
print(f" Average Time to Resolution: {avg_resolution_time:.2f} hours")

# Compare resolution times by priority
priority_resolution = df_cleaned.groupby("Ticket Priority")["Response Time (Hours)"].mean()
print("\n Average Resolution Time by Ticket Priority:\n", priority_resolution)
```

## Pie Chart (Proportion of Tickets by Priority) ##

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set style
sns.set_theme(style="whitegrid")

# Bar plot for average resolution time by ticket priority
plt.figure(figsize=(8, 5))
sns.barplot(x=priority_resolution.index, y=priority_resolution.values, palette="coolwarm")
plt.xlabel("Ticket Priority")
plt.ylabel("Avg Resolution Time (Hours)")
plt.title("Average Resolution Time by Ticket Priority")
plt.show()

# Pie chart for proportion of each ticket priority
plt.figure(figsize=(6, 6))
df_cleaned["Ticket Priority"].value_counts().plot(kind="pie", autopct="%1.1f%%", cmap="coolwarm", startangle=90)
plt.ylabel("")
plt.title("Proportion of Tickets by Priority")
plt.show()
```

The bar chart shows that Critical and Medium-priority tickets are resolved slightly faster (~7 hours) than High and Low-priority tickets (~8 hours), which is unexpected since Critical issues should be addressed the quickest. The pie chart reveals a balanced ticket distribution, with Critical tickets making up the highest proportion (26.7%). This suggests that a high volume of urgent tickets may be affecting resolution efficiency. To improve, the system may need better prioritization to ensure truly critical issues are handled faster.

The average response time has steadily decreased from ~12 to ~8 hours, indicating improved efficiency. This consistent downward trend suggests better workflow management, resource allocation, or automation. Faster responses likely enhance customer satisfaction and issue resolution, boosting overall productivity. If sustained, this trend can lead to higher efficiency and better service quality.
