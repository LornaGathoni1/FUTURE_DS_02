```python
import pandas as pd
# Define file path
file_path = r"C:\Users\user\Desktop\Future Interns\Customer_ticket_ resolution\customer_support_tickets.csv"

# Load dataset without unnecessary unnamed columns
df = pd.read_csv(file_path)


# Display first few rows
df.head()
```

```python
column_names = df.columns.tolist()
print(column_names)
```

```python
df.shape
```

```python
# Check for duplicate rows
duplicates = df.duplicated().sum()
print(f"Number of duplicate rows: {duplicates}")

# Drop duplicates if any
if duplicates > 0:
    df = df.drop_duplicates()
    print("Duplicates dropped.")
```

```python
# Check for missing values
missing_values = df.isnull().sum()
print("\nMissing Values in Each Column:")
print(missing_values)

# Total missing values
total_missing = missing_values.sum()
print(f"\nTotal missing values: {total_missing}")
```

```python
# Check unique values in the missing-value columns to see if they contain blanks or NaNs
print("\nUnique values in 'Resolution' column:\n", df["Resolution"].unique())
print("\nUnique values in 'Time to Resolution' column:\n", df["Time to Resolution"].unique())
print("\nUnique values in 'Customer Satisfaction Rating' column:\n", df["Customer Satisfaction Rating"].unique())
print("\nUnique values in 'First Response Time' column:\n", df["First Response Time"].unique())

# Count non-missing values in each of the 4 columns
non_missing_counts = df[["Resolution", "Time to Resolution", "Customer Satisfaction Rating", "First Response Time"]].count()

# Print the counts of non-missing entries
print("\nNon-Missing Entries in Each Column:\n", non_missing_counts)
```

```python
# Filter rows where at least one of the four columns has missing values
missing_rows = df[df[["Resolution", "Time to Resolution", "Customer Satisfaction Rating", "First Response Time"]].isna().any(axis=1)]

# Count how many such rows exist
num_missing_rows = missing_rows.shape[0]

# Display the first few rows with missing values
print(f"Total rows with missing values in at least one column: {num_missing_rows}")
missing_rows.head()
```

```python
# Separate numerical and categorical columns
numerical_cols = df.select_dtypes(include=['number']).columns.tolist()
categorical_cols = df.select_dtypes(exclude=['number']).columns.tolist()

print("Numerical Columns:", numerical_cols)
print("Categorical Columns:", categorical_cols)
```

```python
columns_to_drop = ["Customer Name", "Customer Email", "Ticket ID", "Customer Age", "Customer Gender", "Date of Purchase"]
df.drop(columns=columns_to_drop, inplace=True)

# Confirm changes
print("Remaining Columns:", df.columns.tolist())
```

```python
import pandas as pd

# Ensure necessary columns exist before processing
if 'Time to Resolution' in df.columns and 'First Response Time' in df.columns:

    # Convert to datetime format while handling errors
    df['First Response Time'] = pd.to_datetime(df['First Response Time'], errors='coerce')
    df['Time to Resolution'] = pd.to_datetime(df['Time to Resolution'], errors='coerce')

    # Calculate Resolution Time in hours
    df['Resolution Time (Hours)'] = (df['Time to Resolution'] - df['First Response Time']).dt.total_seconds() / 3600

    # Handle missing values in "Resolution Time (Hours)"
    df['Resolution Time (Hours)'] = df['Resolution Time (Hours)'].fillna(0).round(2)
    df['Resolution Time (Hours)'] = df['Resolution Time (Hours)'].apply(lambda x: max(x, 0))

    # Fill missing values in datetime columns with descriptive text
    df["First Response Time"] = df["First Response Time"].fillna("Not recorded")
    df["Time to Resolution"] = df["Time to Resolution"].fillna("Not resolved")

# Fill missing values in categorical columns
df["Customer Satisfaction Rating"] = df["Customer Satisfaction Rating"].fillna(df["Customer Satisfaction Rating"].median())
df["Resolution"] = df["Resolution"].fillna("No resolution provided")

# Create a cleaned copy
df_cleaned = df.copy()

# Confirm missing values are handled
print(df_cleaned.isnull().sum())

# Display cleaned dataset preview
df_cleaned.head()
```

```python
import nltk
from collections import Counter
from nltk.corpus import stopwords
from sklearn.feature_extraction.text import CountVectorizer

# Ensure NLTK stopwords are downloaded
nltk.download('stopwords')

# Count occurrences of ticket subjects
subject_counts = df_cleaned['Ticket Subject'].value_counts()
print("Most common ticket subjects:\n", subject_counts.head(10))

# Tokenize and clean ticket descriptions
stop_words = set(stopwords.words('english'))
vectorizer = CountVectorizer(stop_words='english', max_features=20)
desc_matrix = vectorizer.fit_transform(df_cleaned['Ticket Description'])
desc_counts = dict(zip(vectorizer.get_feature_names_out(), desc_matrix.sum(axis=0).A1))

# Display most common words in ticket descriptions
common_words = sorted(desc_counts.items(), key=lambda x: x[1], reverse=True)
print("\nMost common words in descriptions:\n", common_words[:10])
```

```python
import matplotlib.pyplot as plt
from wordcloud import WordCloud, STOPWORDS

# Combine all ticket descriptions into a single text
text = " ".join(df_cleaned['Ticket Description'])

# Define stopwords
stopwords = set(STOPWORDS)

# Generate word cloud
wordcloud = WordCloud(width=800, height=400, background_color='white', stopwords=stopwords, colormap='viridis').generate(text)

# Display the word cloud
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')  # Hide axes
plt.title("Most Common Words in Ticket Descriptions", fontsize=14)
plt.show()
```

```python
# Resolution Time Analysis
import seaborn as sns

# Plot distribution of resolution time
plt.figure(figsize=(8,5))
sns.histplot(df_cleaned['Resolution Time (Hours)'], bins=30, kde=True)
plt.title("Distribution of Resolution Time (Hours)")
plt.xlabel("Resolution Time (Hours)")
plt.ylabel("Frequency")
plt.show()

# Check average resolution time per ticket type
avg_resolution_time = df_cleaned.groupby('Ticket Type')['Resolution Time (Hours)'].mean().sort_values(ascending=False)
print("Average Resolution Time by Ticket Type:\n", avg_resolution_time)
```

Refund requests take the longest to resolve (1.41 hours) due to verification steps, followed by technical issues (1.29 hours) requiring troubleshooting. Cancellation requests, product inquiries (1.20 hours), and billing inquiries (1.17 hours) are faster, likely due to streamlined processes.

```python
#Customer Satisfaction vs. Resolution Time
plt.figure(figsize=(8,5))
sns.scatterplot(x=df_cleaned['Resolution Time (Hours)'], y=df_cleaned['Customer Satisfaction Rating'], alpha=0.5)
plt.title("Resolution Time vs Customer Satisfaction")
plt.xlabel("Resolution Time (Hours)")
plt.ylabel("Customer Satisfaction Rating")
plt.show()
```

atisfaction scores vary across all resolution times, suggesting speed alone doesnt determine customer satisfaction. High ratings (5 stars) still appear for long wait times (10+ hours), indicating good service matters. Low ratings (1-2 stars) exist even for quick resolutions (~05 hours), possibly due to unresolved issues.
